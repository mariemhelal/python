```python
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
import plotly.figure_factory as ff
import seaborn as sns
import plotly.io as pio
pio.templates.default = "plotly_white"
```

```python
import pandas as pd
```

```python
Data = pd.read_csv("customer_acquisition_cost_dataset.csv")
```

```python
Data.head()
```

```python
Data.info()
```

```python
Data ['CAC'] = Data ['Marketing_Spend']  / Data['New_Customers']
Data ['CONV'] = Data ['New_Customers'] / Data['Marketing_Spend'] * 100

Data.head()
```

```python
Fig1= px.scatter(x = Data['New_Customers'], y = Data['Marketing_Spend'] , title = 'Marketing Spend vs New Customers')
Fig1.show()
```

```python

Fig2 = px.funnel (
    data_frame=Data,
    y='Marketing_Channel',
    x='CONV',
    title='Channel vs Conversion Rate',
    orientation='h',
    color_discrete_sequence=['red']
)

Fig2.show()
```
