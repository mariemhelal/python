```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from wordcloud import WordCloud
```

```python
file_path = '/content/amazon.csv.zip'
file_name = 'amazon.csv'
df = pd.read_csv(file_path)
df.head()

```

```python
df.describe()
```

```python
df['rating']= pd.to_numeric(df['rating'], errors ='coerce')
category_per_rating = df.groupby('category').filter(lambda product : product['rating'].sum() >= 100)
category_per_rating.head()
```

```python
category_per_product = df.groupby('category')['product_name'].size().sort_values(ascending = False )
filter_category = category_per_product[category_per_product >= 50]
filter_category.head()
```

```python
df.info()
```

```python
# Use to_numeric and astype to convert to integer, handling NaN values with fillna
rating = pd.to_numeric(df['rating'], errors='coerce').fillna(0)
df['rating']= rating.astype('int64')
discounted_price = pd.to_numeric(df['discounted_price'], errors='coerce').fillna(0)
df['discounted_price']= discounted_price.astype('int64')
df.info()
```

```python
# Use to_numeric and astype to convert to integer, handling NaN values with fillna
import matplotlib.pyplot as plt
from wordcloud import WordCloud

text = ' '.join(df['category'].astype(str).tolist())

wc = WordCloud(width=800, height=400, background_color="white").generate(text)

# Display the word cloud
plt.figure(figsize=(10, 5))
# Change wordcloud to wc
plt.imshow(wc, interpolation="bilinear") # Changed 'wordcloud' to 'wc' to use the generated word cloud
plt.axis("off")
plt.title("Word Cloud for Categories with Higher Price and User ID")
plt.show()
```
